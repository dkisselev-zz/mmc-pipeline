```python
# Step 4: The main code to search PubMed and retrieve DOIs

# It's good practice to provide a tool name and your email to the NCBI API
# You can leave these as is, but it's better to change them.
pubmed = PubMed(tool="MyDOI-Finder", email="dmitry756@gmail.com")

# Split the multi-line string of titles into a Python list
# .strip() removes leading/trailing whitespace, and we filter out any empty lines.
titles = [title.strip() for title in titles_input.strip().split('\n') if title.strip()]

print(f"Found {len(titles)} titles to search for.\n")

# Create a list to store results
results_list = []

# Import time for rate limiting
import time

# Loop through each title in the list
# Process all titles instead of just the first 10
for title in titles:
    # Construct a specific query to search for the title in the [Title] field
    # Using [ti] or [title] makes the search much more accurate.
    # We also ask for a maximum of 5 results in case the top hit isn't perfect.
    query = f'{title}[ti]'

    # Execute the query
    results = pubmed.query(query, max_results=5)

    print(f"Searching for: '{title}'")

    found_doi = None

    # The 'results' object is an iterator. We loop through the articles found.
    for article in results:
        # The article object has a 'doi' attribute
        # We take the first one we find that is not None.
        if article.doi:
            found_doi = article.doi.split()[0] # Take only the first DOI if multiple are present
            break # Stop searching once we've found a DOI for this title

    # Add the result to the list
    results_list.append({'Original Title': title, 'DOI': found_doi})

    # Print the result
    if found_doi:
        print(f"  -> DOI: {found_doi}\n")
    else:
        print(f"  -> DOI not found.\n")

    # Add a small delay to avoid hitting API limits
    time.sleep(0.5) # Adjust delay as needed

print("--- Search Complete ---")

# Step 5: Create a pandas DataFrame from the results
import pandas as pd
df = pd.DataFrame(results_list)

# Step 6: Export the DataFrame to a CSV file
df.to_csv('pubmed_dois.csv', index=False)

print("Results exported to pubmed_dois.csv")
```
